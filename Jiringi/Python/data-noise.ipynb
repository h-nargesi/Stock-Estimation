```python
import codes.handlers as hd

rides = hd.ReadPanda('graph/instruments-types')
print("Records:", rides.shape)
if rides.size > 0:
    rides.plot(title='Intrument Types', x='TitleEn', y='RowCounts', kind='bar', figsize=(20, 5))
```

```python
import codes.handlers as hd

rides = hd.ReadPanda('graph/instrument-scores', ("Score", 30, 220))
print("Records:", rides.shape)
if rides.size > 0:
    rides.plot(title='Intrument Score', x='Instrument', y='Value', kind='bar', figsize=(20, 5))
```

```python
import codes.handlers as hd

rides = hd.ReadPanda('graph/density', ("19835, 19823, 19843", 10,))
print("Records:", rides.shape)
if rides.size > 0:
    rides = rides.pivot(index='DateTimeEn', columns='Instrument', values=['Density'])
    # print(rides.columns)
    rides.plot(title='Density', kind='line', figsize=(20, 5))
```

```python
import codes.handlers as hd

rides = hd.ReadPanda('graph/trade', ("19837, 19835, 19823, 19843", 6,))
print("Records:", rides.shape)
if rides.size > 0:
    # 'OpenPrice', 'LowPrice', 'HighPrice', 
    rides = rides.pivot(index='DateTimeEn', columns='Instrument', values=['ClosePrice'])
    # print(rides.columns)
    rides.plot(title='Trade', kind='line', figsize=(20, 5))
```
