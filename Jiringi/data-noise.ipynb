```python
from codes.handlers import Handlers

hd = Handlers('queries')
rides = hd.ReadPanda('graph/density', (None, 40, "2014",))
print("Records:", rides.shape)
if rides.size > 0:
    rides = rides.pivot(index='DateTimeEn', columns='Instrument', values=['Density'])
    # print(rides.columns)
    rides.plot(title='Density', kind='line', figsize=(20, 5))
```
